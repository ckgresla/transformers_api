```python
# Testing Parse Function

keyphrase_tokens = []
keyphrases = [(i["word"], i["entity"][0], i["start"], i["end"]) for i in out] #get relevant portions of output

def parse_keywords(self, keyphrases):
    for word, label, start, end in keyphrases:
        if label == "B":
            keyphrase_tokens.append([word])
            prev_end = end #for checking next span
        elif label == "I":
            if len(keyphrase_tokens) > 0:
                # Get Consecutive Spans together (changes outputs from [key, phrase, extractor] into [keyphrase, extractor])
                if prev_end == start:
                    keyphrase_tokens[len(keyphrase_tokens) - 1][-1] += word #if current KeyPhrase  span is part of last word
                else:
                    keyphrase_tokens[len(keyphrase_tokens) - 1].append(word) #appends new word to last list of KeyPhrase Spans

    return keyphrase_tokens
```
